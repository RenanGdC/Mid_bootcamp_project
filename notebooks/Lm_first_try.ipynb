```python
import pandas as pd
import numpy as np
```

```python
employee_df = pd.read_excel('../data/HR_Employee_Data1.xlsx')
employee_df.head()
```

Preparing data for the prediction model:

```python
employee_df = employee_df.drop(['Unnamed: 0'], axis=1)
```

```python
employee_df['salary'] = employee_df['salary'].replace({'low':1, 'medium':2, 'high':3})
```

```python
employee_df.head()
```

```python
department_enc = pd.get_dummies(employee_df['department'])
department_enc
```

Using get_dummies to make the column department numerical for the linear regression and droping emp_id:

```python
employee_df = employee_df.drop(['department'], axis=1)
```

```python
employee_df = pd.concat([employee_df,department_enc], axis = 1)
employee_df.head()
```

```python
employee_df = employee_df.drop(['emp_id'], axis = 1)
employee_df.head()
```

Checking data for nas:

```python
employee_df.isnull().sum()
```

```python
#pip install sklearn.preprocessing
```

```python
X = employee_df.drop(['left'], axis = 1)
y = employee_df['left']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled_df = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_scaled_df
```

Spliting the data in order to compare the train and test results:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size = 0.3, random_state=66)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Applying the linear regression:

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred_train = lm.predict(X_train)
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = lm.predict(X_test)
r2_score(y_test, y_pred_test)
```

The result error is way to high in comparison to the expected.
