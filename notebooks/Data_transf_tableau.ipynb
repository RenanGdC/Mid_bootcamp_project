```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
employee_df = pd.read_excel('../data/HR_Employee_Data1.xlsx')
employee_df.head()
```

```python
employee_df = employee_df.drop(['Unnamed: 0'], axis=1)
```

```python
employee_df['left'] = employee_df['left'].replace({1:'Yes', 0:'No'})
```

```python
employee_df.head()
```

```python
bin_edgamh=[96,150,200,250,310]
bin_lblamh=['<150', '150-200', '200-250', '250-310']
employee_df['average_montly_hours_lbl']=pd.cut(employee_df['average_montly_hours'], bins = bin_edgamh,
                                               labels = bin_lblamh)
sns.countplot(x=employee_df['average_montly_hours_lbl'], hue=employee_df["left"]).set(
     xlabel="Avg. Monthly Hours")
plt.show()
```

```python
employee_df['satisfaction_level'].describe()
```

```python
bin_edgamh=[0,0.35,0.5,0.75,1]
bin_lblamh=['Low', 'Avg Low', 'Avg High', 'High']
employee_df['satisfaction']=pd.cut(employee_df['satisfaction_level'], bins = bin_edgamh,
                                               labels = bin_lblamh)
```

```python
employee_df.to_excel("../data/HR_Employee_Data_tableau2.xlsx")
```
