```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
employee_df = pd.read_excel('../data/HR_Employee_Data1.xlsx')
employee_df.head()
```

```python
employee_df = employee_df.drop(['Unnamed: 0'], axis=1)
```

```python
employee_df['salary'] = employee_df['salary'].replace({'low':1, 'medium':2, 'high':3})
employee_df.head()
```

Droping columns:

```python
employee_df = employee_df.drop(['emp_id'],axis=1)
employee_df = employee_df.drop(['last_evaluation'], axis=1)
employee_df = employee_df.drop(['number_project'], axis=1)
```

Encoding the department column:

```python
department_enc = pd.get_dummies(employee_df['department'])
department_enc
```

Concatenating the encoded departments df to our main and droping the department column:

```python
employee_df = employee_df.drop(['department'], axis=1)
employee_df = pd.concat([employee_df,department_enc], axis=1)
employee_df.head()
```

Applying the scaler:

```python
X = employee_df.drop(['left'],axis=1)
y = employee_df['left']
```

```python
scaler = MinMaxScaler()
X_scaled_df = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_scaled_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=66)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Applying the regression model:

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred_train=lm.predict(X_train)
r2_score(y_train,y_pred_train)
```

```python
y_pred_test=lm.predict(X_test)
r2_score(y_test,y_pred_test)
```
